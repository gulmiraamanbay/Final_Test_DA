```python
 Задание 8
Во вкладке "Данные АБ-тестов" результаты трех несвязанных АБ тестов для ARPU (общая выручка/общее количество пользователей).
Посмотрите на результаты тестов и интерпретируйте их. Напишите значения p-value, которые вы получили.
Подготовьте выводы и рекомендации. 

experiment_num - номер эксперимента
experiment_group - группа, в которую попал пользователь
user_id - id пользователя
revenue - выручка, которую сгенерировал пользователь, купив платную услугу продвижения
Ваш ответ:
```

```python
Данные для тестового задания
```

```python
import pandas as pd
from scipy import stats

# 1. Загружаем данные (укажите путь к вашему файлу)
df = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные АБ тестов')

# Пройдемся циклом по всем трем экспериментам
for i in [1, 2, 3]:
    exp_data = df[df['experiment_num'] == i]
    
    test = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    control = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    
    # Считаем ARPU (среднее)
    arpu_test = test.mean()
    arpu_control = control.mean()
    
    # Проводим Т-тест (Уэлча, так как дисперсии могут быть разными)
    t_stat, p_val = stats.ttest_ind(test, control, equal_var=False)
    
    print(f"--- Эксперимент №{i} ---")
    print(f"ARPU Test: {arpu_test:.2f} | ARPU Control: {arpu_control:.2f}")
    print(f"P-value: {p_val:.4f}")
    
    if p_val < 0.05:
        print("Результат: СТАТИСТИЧЕСКИ ЗНАЧИМ. Рекомендуется внедрять.")
    else:
        print("Результат: НЕ ЗНАЧИМ. Разница случайна.")
    print("\n")
```

```python
 Задание 9 По датасету с листерами посчитайте средний доход на пользователя 
```

```python
df_listers = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Листеры')

# 2. Считаем общую выручку
total_revenue = df_listers['revenue'].sum()

# 3. Считаем количество УНИКАЛЬНЫХ пользователей (листеров)
# Это важно, так как один ID может встречаться многократно
unique_listers = df_listers['user_id'].nunique()

# 4. Считаем ARPU (средний доход на одного пользователя)
arpu_listers = total_revenue / unique_listers

print(f"Общая выручка: {total_revenue:.2f}")
print(f"Количество уникальных листеров: {unique_listers}")
print(f"Средний доход на пользователя (ARPU): {arpu_listers:.2f}")
```

```python
# 1. Оставляем только уникальных пользователей, чтобы каждый возраст посчитался один раз
unique_users_age = df_listers.drop_duplicates(subset='user_id')['age']

# 2. Считаем медиану
median_age = unique_users_age.median()

print(f"Медианный возраст пользователя: {median_age}")
```

```python
Задание 18  Были получены следующие результаты. Коллеги просят вас подтвердить их и сделать окончательный вывод по эксперименту.
•	Вариант A (контрольная группа) — 100 047 501 посетитель, 1003 платежа.
•	Вариант B (тестовая группа) — 100 001 055 посетителей, 1099 платежей.
Какие рекомендации вы бы дали, основываясь на этих данных?
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Данные эксперимента
count = np.array([1003, 1099])       # Количество платежей (A и B)
nobs = np.array([100047501, 100001055])  # Количество посетителей (A и B)

# Проводим Z-тест
z_stat, p_value = proportions_ztest(count, nobs)

# Расчет конверсии в процентах
cr_a = (count[0] / nobs[0]) * 100
cr_b = (count[1] / nobs[1]) * 100
lift = (cr_b - cr_a) / cr_a * 100

print(f"Конверсия А: {cr_a:.6f}%")
print(f"Конверсия B: {cr_b:.6f}%")
print(f"Относительный прирост: {lift:.2f}%")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Результат: СТАТИСТИЧЕСКИ ЗНАЧИМ. Разница не случайна.")
else:
    print("Результат: НЕ ЗНАЧИМ. Разница может быть шумом.")
```
